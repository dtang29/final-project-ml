# Logistic Regression Prediction Model, or A Lesson in the Pitfalls of Data

## We run a logistic regression in order to predict whether a restaurant location is likely to remain open or close, defined as the encoded discrete variables "Yes/No".  

```python
import seaborn as sns
import pandas as pd
import numpy as np

import pylab as pl
import numpy as np
```

### We import our dataset for restaurants that opened in San Francisco after 2010; with data being defined at the unique location of each restaurant.

```python
df_csv = pd.read_csv('cleanDatasets/master_dataset2010.csv')
df = pd.DataFrame(df_csv)
df_reg = df[[
    'rating','isClosed','lic_code_cat'
]].dropna()

print("A Brief Summary of our variables:")
print(" ")
print(" ")
print("isClosed = 1 indicates a 'No', meaning a restaurant is open; a 0, i.e. a 'Yes', means a location is closed")
print(str(len(df_reg[df_reg['isClosed']==0])/len(df_reg)) + "% of the dataset is made up of closed locations")
print(" ")
print("lic_code_cat refers to the encoded registered business categories as conventionally defined by the city of San Francisco: -1 is Null, 0 refers to 'Fast Food Extablishment', 1 refers to 'Restaurant 1,000 - 2,000 Sqft', 2 refers to 'Restaurant Greater Than 2,000 Sqft', 3 refers to 'Restaurant Less Than 1,000 Sqft', and 4 refers to 'Take-Out Establishment' ")
print(" ")
print('rating are user generated by users on Google Maps')
print(" ")
print("n = " + str(len(df_reg)))
print(" ")
print(" ")
print(df_reg.head())
# plot all of the columns
print(" ")
df_reg.hist()
pl.show()
```

```python
print("We define our features variables as X, and our target as y")
X = df_reg.drop("isClosed", axis=1)
y = df_reg["isClosed"]
print(X.shape, y.shape)
```

#### We fit our data into a Logistic Regression Model using the statsmodels package.

```python
import statsmodels.api as sm
logit_model=sm.Logit(y,X)
result=logit_model.fit()
print(result.summary())
print("")
print("Confidence Interval")
print(result.conf_int())
```

We can infer from the regression summary results that the coefficients for our explanatory variables is significant. However the question of whether they explain a significant amount of the variability in restaurants remaing closed or open is a different question altogether.

The Confidence Interval indicate that there is no inverse relationship between our Xs and whether a restaurant is likely to close or remain open. Note that the C.I. is larger in predicting whether a location is likely to remain open (1) than closed (0). This is likely due to the ratio of open to closed locations in our data.

#### We run the Machine Learning prediction model using sklearn

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)

logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

Despite the high accuracy score, something is amiss. One cannot help but think that the model is deeply flawed, as no "Closed" locations are trained nor predicted. If the model can only predict a single category, then it is truly useless and utterly fails to provide valuable insights.

### Refining the Model

##### There are multiple considerations to ponder in refining the model, namely those are:
   
   1) Whether our sample is truly representative: we happen to only have data for closed location from 2015 onward, and yet we have included data from 2010 onward. This posits that no restaurants in San Francisco closed from 2010-2015, and we know well that is highly unlikely, and that our data source was thus incomplete. Therefore, we need to filter our data from 2015 onward so that we have a better representation of the ratio of closed and open restaurants.
   
   2) Filtering out NULL values: sometime mistakes happen, even during the cleanup process. It would seem that NULL values for lic_code were encoded and thus accounted for as a "meaningful" categorical variable. The histogram of the distribution of lic_code indicate a high abundance of NULLs. Since NULLs do not provide contextual understanding to predicting open or closed location, it would be wiser to remove those entirely. However this entails dropping the representation of "Closed" data from 20 to 3%.  
   
   3) Filtering values that are meaningful: the lic_code variable has 3 classes pertaining to a restaurant's square footage, while the other two could be described as "types" (fast food, takeaways). There seem to be a contextual conflict, and it would be intuitive to retain square footage as a predictor value.
   
   4) Creating dummy variables: lic_code has been split into 6 categories (including NULLs), which, once filtered by square footage, will be down to 3. This wil enable better control over how those categorical variables are represented, and interpreted in the model.

#### Filtering the data's range for better representation (Consideration 1, 2 & 3)

```python
df_csv = pd.read_csv('cleanDatasets/master_dataset2015.csv')
df = pd.DataFrame(df_csv)

df['lic_code'] = df['lic_code'].fillna("No_data")
df["lic_code"] = df["lic_code"].astype('category')
df['lic_code_cat'] = df['lic_code'].cat.codes

df_reg = df[[
    'rating','isClosed','lic_code_cat'
]].dropna()
len(df_reg)

# df_reg = df_reg.drop(df_reg[df_reg.lic_code_cat.isin(['-1'])].index)
df_reg = df_reg[df_reg.lic_code_cat != 4]
df_reg = df_reg[df_reg.lic_code_cat != 5]

print("A Brief Summary of our variables:")
print(" ")
print("isClosed = 1 indicates a 'No', meaning a restaurant is open; a 0, i.e. a 'Yes', means a location is closed")
print(" ")
print(str(len(df_reg[df_reg['isClosed']==0])/len(df_reg)) + "% of the dataset is made up of closed locations")
print(" ")
print("lic_code_cat is now filtered down to: 0 refers to NULLs, 1 refers to 'Restaurant 1,000 - 2,000 Sqft', 2 refers to 'Restaurant Greater Than 2,000 Sqft', and 3 refers to 'Restaurant Less Than 1,000 Sqft' ")
print(" ")
print("n = " + str(len(df_reg)))
print(" ")
print(df_reg.head())
# plot all of the columns
print(" ")
df_reg.hist()
pl.show()
```

#### Creating dummy variables for lic_code_cat (Consideration 4)

```python
# dummify rank
dummy_ranks = pd.get_dummies(df_reg['lic_code_cat'], prefix='lic_code_cat')
print(dummy_ranks.head())
print('')

# create a clean data frame for the regression
cols_to_keep = ['isClosed', 'rating']
data = df_reg[cols_to_keep].join(dummy_ranks.loc[:, 'lic_code_cat_1':])
print(data.head())

# manually add the intercept
data['intercept'] = 1.0
```

```python
train_cols = data.columns[1:]

logit = sm.Logit(data['isClosed'], data[train_cols])

# fit the model
result = logit.fit()

print(result.summary())
print("")
print("Confidence Interval")
print(result.conf_int())
print("")
print("Odds Ratio")
print(np.exp(result.params))
```

Rating is no longer statistically significant. And while the Odds Ratio seems to indicate that lic_code_cat_3 (i.e. 'Restaurant Less Than 1,000 Sqft') have greater odds to remain open compared to the two other categories, it is hard to infer how significant of a correlation in predicting whether a restaurant is likely to stay open or close. 

```python
X = data.drop("isClosed", axis=1)
y = data["isClosed"]
print(X.shape, y.shape)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=0)
from sklearn.linear_model import LogisticRegression
from sklearn import metrics
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
```

```python
y_pred = logreg.predict(X_test)
print('Accuracy of logistic regression classifier on test set: {:.2f}'.format(logreg.score(X_test, y_test)))
```

```python
from sklearn import model_selection
from sklearn.model_selection import cross_val_score
kfold = model_selection.KFold(n_splits=10, random_state=7)
modelCV = LogisticRegression()
scoring = 'accuracy'
results = model_selection.cross_val_score(modelCV, X_train, y_train, cv=kfold, scoring=scoring)
print("10-fold cross validation average accuracy: %.3f" % (results.mean()))
```

```python
from sklearn.metrics import confusion_matrix
confusion_matrix = confusion_matrix(y_test, y_pred)
print(confusion_matrix)
```

The matrix is telling us that we have 0+73 correct predictions and 0+20 incorrect predictions; note that it is not predicting any closed locations at all, because I suspect the training data was not sufficient to generate training points for Closed locations. By dropping NULLs, the representation of closed restaurant dropped from roughly 18% of the dataset to a mere 3%. If we are to predict margins this thin with such a sample representation, we need either better explanatory variables or large samples of the ones we have (although most likely with similar limited outcomes).

```python
from sklearn.metrics import classification_report
print(classification_report(y_test, y_pred))
```

```python
import matplotlib.pyplot as plt 
plt.rc("font", size=14)

from sklearn.metrics import roc_auc_score
from sklearn.metrics import roc_curve
logit_roc_auc = roc_auc_score(y_test, logreg.predict(X_test))
fpr, tpr, thresholds = roc_curve(y_test, logreg.predict_proba(X_test)[:,1])
plt.figure()
plt.plot(fpr, tpr, label='Logistic Regression (area = %0.2f)' % logit_roc_auc)
plt.plot([0, 1], [0, 1],'r--')
plt.xlim([0.0, 1.0])
plt.ylim([0.0, 1.05])
plt.xlabel('False Positive Rate')
plt.ylabel('True Positive Rate')
plt.title('Receiver operating characteristic')
plt.legend(loc="lower right")
plt.savefig('Log_ROC')
plt.show()
```

The ROC (Receiver Operating Characteristic) curve indicates clearly the shortcoming of our outcome variables; the logistic regression curve, if it were to model a proper predictive model, would be, well, curved and far from the dotted line as possible. However the proximity and lack of curvature from 0, i.e. the closed category, further highlights how inefficient this model is.

As a result it also seems that we are overfitting the prediction about locations staying open.

Thus the predictive models are proven to be inconclusive. For the model to yield value it needs to be able to predict with reasonable accuracy (> 80%) whether a location is to remain open or close. While we derived a better closure to opening ratio by filtering the temporal range of our dataset, having an abundance of NULLs in our categorical variable creates noise  and skews greatly the output of our model.

## Thoughts and Conclusion

It is abundantly clear that we have two major issues at hand:
    1) Pertinence of categorical data;
    2) Thoroughness of data;
    3) Appropriate representation of the outcome variable.
    
Simply put, the current dataset was not ready for the model. However we have demonstrated a paradox of statistical thinking: that models' successes do not merely depend on whether the coefficients of input variables are statistically significant, nor on the accuracy score of prediction, as those are simply a means to an end. Rather that it is in recontextualizing statistical outputs to the real-world problematic our model seeks to remedy that we gain a better understanding of data, and in addressing its flaws.
